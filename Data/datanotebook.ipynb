```python
def scrape_team_info(team,sport):
    # Send a GET request to the URL
    teamname = team
    url = "https://www.osdbsports.com/"+sport.lower() +"/" + team.lower().replace(" ","-")
    response = requests.get(url)
    
    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Parse the HTML content using BeautifulSoup
        soup = BeautifulSoup(response.content, 'html.parser')
        data = soup.find_all("h3",attrs={"class":"text-white text-break"})
        data = [h3.text.strip() for h3 in soup.find_all("h3", attrs={"class": "text-white text-break"})]
        trophies = soup.find_all("div",attrs={"class":"font-weight-bold award-year align-self-start"})
        if sport == "MLS":
            data[1] = data[1].replace("MLS 2024,","")
        else:
            del data[2]

        if len(data[0]) == 9:
            data[0] = data[0][5:9]
        
        team = {"Team":teamname,"Location":data[4],"Championshsips":len(trophies),"League":sport,"Conference":data[1], "Founded":data[0][0:4]}
        stadium ={"Name":data[2],"Size":data[3],"Location":data[4]}
        staff = pd.DataFrame(columns = ["Name","Position","Team"])
        row = {"Name":data[5],"Position":"Owner", "Team": teamname}
        staff.loc[len(staff)] = row
        if (len(data) == 9):
            row = {"Name":data[8],"Position":"Head Coach", "Team": teamname}
            staff.loc[len(staff)] = row
        return stadium, team, staff
    
    else:
        print("Failed to fetch page:", response.status_code)
        return None, None, None
```

```python
def extract_revenues(soup):
    revenues = []
    
    # Find all <strong> tags within <li> tags
    strong_tags = soup.find_all('li')
    
    # Loop through each <strong> tag to find the one containing 'Revenue:'
    for tag in strong_tags:
        if 'Revenue:' in tag.text:
            # Extract the revenue text
            revenue_text = tag.text
            
            # Extract the revenue figure
            revenue = revenue_text.split('$')[1].strip()
            
            revenues.append(revenue)
    
    return revenues

def convert_revenues_to_int(revenues):
    revenue_integers = []
    for revenue_str in revenues:
        # Split the string to separate the number and the unit
        amount, unit = revenue_str.split()
        # Convert the amount to float to handle decimal values
        amount_float = float(amount)
        # Convert the unit to a multiplier
        if unit.lower() == 'million':
            amount_int = int(amount_float * 1_000_000)
        elif unit.lower() == 'billion':
            amount_int = int(amount_float * 1_000_000_000)
        else:
            amount_int = int(amount_float)
        # Append the converted revenue to the list
        revenue_integers.append(amount_int)
    return revenue_integers
```

```python
def teamcolor(t):
    t =  t.replace(' ','_') 
    url = "https://en.wikipedia.org/wiki/"
    url = url + t
    page = requests.get(url)
    soup = BeautifulSoup(page.content, 'html.parser')
    td_text  = soup.find('td', class_="infobox-data").text
    color_names =  re.findall(r'[A-Za-z\s]+', td_text)
    color_names = [name.strip() for name in color_names if name.strip()]
    return color_names

print(teamcolor("Atlanta Hawks"))

```
